**Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("recharge amount must be at least Rs.50")
        return False
    elif amount not in valid_plans:
        print("invalid plan selected")
        print(f"available plans: {valid_plans}")
        return False
    else:
        print("recharge successful")
        return True

while True:
    try:
        amount = int(input("enter recharge amount: Rs."))
        if validate_recharge(amount):
            break
        print("try again")
    except ValueError:
        print("enter a valid numeric amount")
```

**Inventory Reorder Alert System**

```python
products = {
    'Mobile': 15,
    'Laptop': 30,
    'Headphones':8,
    'Smart TV': 5
}

def check_stock(product, quantity):
    if quantity < 15:
        print(f"{product}: reorder alert")
    else:
        print(f'{product}: stock OK')

for product, quantity in products.items():
    check_stock(product, quantity)
```

**Student Result Processing System**

```python
def calculate_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)
    if average >= 50:
        result = "pass"
    else:
        result = "fail"
    return average, result

student_marks = [35, 48, 70, 85,90]
average, result = calculate_result(student_marks)
print(f"average marks: {average}")
print(f"result: {result}")
```

**Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance* per_km_rate)

    if peak:
        fare *= 1.25
    return fare

while True:
    distance = float(input("enter distance (in km): "))
    peak_hour = input("peak hour? (y/n): ").lower()
    peak = True if peak_hour == 'y' else False

    total_fare = calculate_fare(distance, peak)

    print(f"Total Fare: {total_fare: .2f}")

    retry = input('calculate again? (y/n): ').lower()
    if retry != 'y':
        break
```

**Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance):
    present_days = 0
    
    for day in attendance:
        if day == 'P':
            present_days +=1

    total_days = len(attendance)
    percentage = (present_days / total_days)*100

    if percentage >= 75:
        return percentage, "eligible"
    else:
        return percentage, "not eligible"

attendance_list = ['P','P','A','P','P','A','P','P']

percentage, status = check_attendance(attendance_list)

print(f"attendance percentage: {percentage}")
print(f"eligibility status: {status}")
```

**Password Strength Checker**

```python
def check_password(password):
    if len(password) < 8:
        return "weak password"
    
    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "strong password"
    else:
        return "weak password"

user_password = input("enter password: ")
result = check_password(user_password)
print(result)
```
