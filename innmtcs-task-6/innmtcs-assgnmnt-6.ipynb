**Smart Parking Lot Management System**

```python
def parking_sys(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print(f"currently parked vehicles: {parked}")

    if parked > capacity:
        print("parking status: exceeded capacity")
    else:
        print("parking status: available")

parking_capacity= 50
vehicle_logs= ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_sys(parking_capacity, vehicle_logs)
```

**Online Food Delivery Time Estimator**

```python
def delivery_time(dist, traffic, weather):
    eta = dist * 5

    match traffic:
        case "Low":
            eta+=1
        case "Medium":
            eta+=5
        case "High":
            eta+=10

    match weather:
        case "Cloudy":
            eta+=1
        case "Rainy":
            eta+=5
        case "Smoggy":
            eta+=10

    print(f"estimated delivery time: {eta} minutes")

distance= 8
traffic_level="High"
weather= "Rainy"

delivery_time(distance,traffic_level,weather)
```

**Movie Theatre Seat Occupancy**

```python
def movi_seats(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        message="housefull"
    if occupancy >= 75:
        message="additional shows"
    else:
        message="no additional message"

    print(f"occupancy: {int(occupancy)}%")
    print(f"message: {message}")

total_seats= 200
booked_seats= [1]*150

movi_seats(total_seats, booked_seats)
```

**Cloud Server Load Classification System**

```python
def server_load(cpu):
    ttl = 0

    for usage in cpu:
        ttl += usage    
    avg = ttl / len(cpu)

    if avg < 50:
        status = "normal"
    elif avg < 80:
        status = "warning"
    else:
        status = "critical"

    print(f"average cpu usage: {int(avg)}%")
    print(f"server status: {status}")

cpu_data = [45, 60, 70, 85, 90]

server_load(cpu_data)
```

**Smart Classroom Resource Usage Monitor**

```python
def resource_usage(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert="yes"
    else:
        alert="no"

    print(f"overused resources: {', '.join(overused) if overused else 'none'}")
    print(f"energy alert: {alert}")

resource_usage({
 "Projector": 6,
 "AC": 9,
 "Lights": 4
})
```

**Online Event Registration Capacity Controller**

```python
def event_regist(capacity, regist):
    confirmed = 0
    waitlisted = 0

    for i in range(regist):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if registrations > capacity:
        status = "closed"
    else:
        status = "open"
    
    print(f"confirmed registrations: {confirmed}")
    print(f"waitlisted users: {waitlisted}")
    print(f"registration status: {status}")

event_capacity = 100
registrations = 105

event_regist(event_capacity, registrations)
```
