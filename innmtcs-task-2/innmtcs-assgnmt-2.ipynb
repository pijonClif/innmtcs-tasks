**unique words in a sentance**

```python
sentance="python is easy and python is powerful"

words=set(sentance.split())
print(f"unique words count: {len(words)}")
print(f"unique words: {words}")
```

**highest salary from employee data**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

hi_sal=0
hi_sal_emp=""

for emp_name, sal in employees.items():
    if(sal>hi_sal):
        hi_sal=sal
        hi_sal_emp=emp_name
print(f"highest salary: {hi_sal_emp} - {hi_sal}")
```

**max and min values**

```python
numbers = [45, 22, 89, 10, 66]

max=numbers[0]
min=numbers[0]

for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num
print(f"list: {numbers}")
print(f"max: {max} \nmin: {min}")
```

**count products above a price threshold**

```python
prices = [450, 1200, 899, 1500, 300]

count=0
for price in prices:
    if price>1000:
        count+=1
print(f"products above 1000: {count}")
```

**attendance percentage**

```python
attendance = ["P", "P", "A", "P", "P"]

count=0
for a in attendance:
    if a=="P":
        count+=1
ttl=len(attendance)
perc=count/ttl*100
print(f"attendance percentage: {perc}")
```

**remove duplicate phone numbers**

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

uniq_num=set(phone_numbers)
print(f"unique phone numbers: {uniq_num}")
```

**count char freq**

```python
input_text=input("enter text")

count={}
for char in input_text.lower():
    if char in count:
        count[char]+=1
    else:
        count[char]=1
print(f"character frequency: {count}")
```

**avg marks**

```python
marks = [45, 22, 89, 10, 66]

ttl=0
for m in marks:
    ttl+=m
marks_count=len(marks)
avg=ttl/marks_count

print(f"average marks: {avg}")
```
