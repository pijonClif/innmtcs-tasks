**Smart Light Controller**

```python
class SmartLight:
    def __init__(self, lightN):
        self.lightN = lightN
        self.status = "OFF"

    def turnOn(self):
        self.status = "ON"

    def turnOff(self):
        self.status = "OFF"

    def dispStatus(self):
        print(f"{self.lightN} is {self.status}")

light = SmartLight("Bedroom Light")

light.turnOn()
light.dispStatus()
```

**Employee ID Card System**

```python
class Employee:
    def __init__(self, name, empId, dept):
        self.name = name
        self.empId = empId
        self.dept = dept

    def dispId(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.empId}")
        print(f"Department: {self.dept}")

emp = Employee("Rahul", "EMP102", "IT")

emp.dispId()
```

**Mobile Contact Record**

```python
class Contact:
    def __init__(self, name, phoneN):
        self.name = name
        self.phoneN = phoneN

    def dispContact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone Number: {self.phoneN}")

contact1 = Contact("Anita", "9876543210")
contact1.dispContact()
```

**Movie Rating Display System**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def dispMovie(self):
        print("Movie Details")
        print(f"Name: {self.name}")
        print(f"Rating: {str(self.rating)}/5")
    
movie1 = Movie("Inception", 4.8)

movie1.dispMovie()
```

**Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customerName, addr):
        self.customerName = customerName
        self.addr = addr

    def dispDelivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customerName}")
        print(f"Address: {self.addr}")

delivery = Delivery("Suman", "Hyderabad")

delivery.dispDelivery()
```
