**User Login Check**

```python
crt_username="admin"
crt_password="1234"

username=input("enter username: ")
password=input("enter password: ")

if username==crt_username:
    if password==crt_password:
        print("Login Successful")
else:
    print("Invalid Credentials")
```

**Pass/Fail Analyzer**

```python
marks = [45, 78, 90, 33, 60]

pass_cnt=0
fail_cnt=0

for mark in marks:
    if mark>=50:
        pass_cnt=pass_cnt+1
    else:
        fail_cnt=fail_cnt+1

print("total number of students: ", len(marks))
print("number of pass students: ",pass_cnt)
print("number of fail students: ",fail_cnt)
```

**Simple Data Cleaner**

```python
names = [" Alice ", "bob", " CHARLIE "]
cln_names=[]

for name in names:
    name=name.strip()
    name=name.lower()
    cln_names.append(name)

print("cleaned names: ",cln_names)
```

**Message Length Analyser**

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print("message: ",msg)
    print("length: ",len(msg))
    if len(msg)>10:
        print("*this message is longer than 10 characters*")
    print()
```

**Error Message Detector**

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
err_logs=0

for log in logs:
    if log=="ERROR":
        err_logs=err_logs+1
print("total error count: ",err_logs)
```
