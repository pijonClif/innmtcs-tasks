**search query keyword analysis**

```python
import string
query="Buy mobile phone buy phone online"

stmnt=query.lower()
for char in string.punctuation:
    stmnt=stmnt.replace(char, "")

words=stmnt.split()
freq={}
for word in words:
    freq[word] = freq.get(word, 0) + 1

disp_search={}
for word, count in freq.items():
    if count > 1:
        disp_search[word]=freq.get(word, count)

print(disp_search)
```

**sensor data validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings=[]
for index, value in enumerate(sensor_readings):
    if value%2==0:
        valid_readings.append((index,value))
print(valid_readings)
```

**email domain usage analysis**

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count={}
for email in  emails:
    domain=email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
ttl_emails=len(emails)

for domain,count in domain_count.items():
    perc=(count/ttl_emails)*100
    print(f"{domain}: {int(perc)}%")
```

**sales spike detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales=sum(sales)/len(sales)
threshold=avg_sales*1.29

for index, sale in enumerate(sales):
    if sale >= threshold:
        print(f"day {index+1}: {sale}")
```

**duplicate user id detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
